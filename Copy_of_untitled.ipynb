```python
import numpy as np
import cv2
from glob import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from google.colab.patches import cv2_imshow
```

```python
femail_photos = glob('/content/drive/MyDrive/FaceRecognition/OpenCV/data/female/*.jpg')
```

```python
man_photos = glob('/content/drive/MyDrive/FaceRecognition/OpenCV/data/male/*.jpg')
```

```python
face_model =cv2.CascadeClassifier('/content/drive/MyDrive/FaceRecognition/OpenCV/model/haarcascade_frontalface_default.xml')
```

##**Female images **

```python
img =cv2.imread(femail_photos[1])
image = img.copy()
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
face_rect = face_model.detectMultiScale(gray,1.3,5)
for x,y,w,h in face_rect:
    cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)
cv2_imshow(image)
```

```python
face_model =cv2.CascadeClassifier('/content/drive/MyDrive/FaceRecognition/OpenCV/model/haarcascade_frontalface_default.xml')
for i in range(len(femail_photos)):
    img =cv2.imread(femail_photos[i])
    img_rgb= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    image = img.copy()
    gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    face_list=face_model.detectMultiScale(gray, 1.3 , 5)

    for x, y, w, h in face_list:

      cord=img_rgb[y:y+h, x:x+w]
      cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)
      cut=image[y:y+h, x:x+w]
      cv2.imwrite(f'/content/drive/MyDrive/FaceRecognition/OpenCV/data/save_pictures/female/female_photo{i}.jpg', cut)
```
